```python
#TO-DO LIST
# Sample User Database
user_database = {"shreedhar": {"name": "Shreedhar", "age": 18, "email": "shreedhar@gmail.com", "password": "12345"},
                 "amit10": {"name": "Amit", "age": 20, "email": "amit@gmail.com", "password": "abcdef"}}

# Function to handle user login
def login():
    username = input("Enter your username: ")
    password = input("Enter your password: ")
    if username in user_database and user_database[username]["password"] == password: # Here I used simple decision making sentence along with the implementation of membership operator and key values of dictionary
        print("Login successful!")
        return user_database[username]
    else:
        print("Invalid username or password")
        return None
# Function to view user information
def view_profile(user):
    print("Name:", user["name"])
    print("Age:", user["age"])
    print("Email:", user["email"])

# Function to update user profile
def update_profile(user):
    print("Update Profile:")
    user["name"] = input("Enter new name: ")
    user["age"] = input("Enter new age: ")
    user["email"] = input("Enter new email: ")
    print("Profile updated successfully!")

# Main menu function
def main_menu():
    while True: # Using while loop I simply created a Menu-Driven approach without using match case
        print("\nMenu:")
        print("1. Login")
        print("2. View Profile")
        print("3. Update Profile")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            user = login() # It will call the login() function and if the username and password is correct it will store the detail of the username(key) into the variable user
            if user:
                todo_list(user)
        elif choice == "2":
            print("\nView Profile:")
            user = login() # It will call the login() function and if the username and password is correct it will run the view_profile() function where we can see the details of the user
            if user:
                view_profile(user)
        elif choice == "3":
            print("\nUpdate Profile:")
            user = login() # It will call the login() function and if the username and password is correct it will run the update_profile() function where we can update the values of the keys i.e. Name, Age, Mail
            if user:
                update_profile(user)
        elif choice == "4":
            print("Exiting program...")
            break # Simply terminates the code
        else:
            print("Invalid choice, please try again.")

# To-do menu function
def todo_list(user):
    tasks={}
    while True:
        print("\nTO-DO List:")
        print("1. Add Task")
        print("2. View Task")
        print("3. Mark as Done")
        print("4. Logout")
        choice = input("Enter your choice: ")

        if choice == "1": # It will ask the user to enter number of task and will store them one by one
            num=int(input("Enter number of tasks : "))
            for i in range(num):
                task=input("Enter Task : ")
                tasks[task]="Not Done"
            print("Tasks Added...")
        elif choice == "2": # It will print the task(value) stored in the dictionary variable tasks along with the value of the key i.e. the status of the task
            print("\nTasks")
            for key in tasks:
                print(f"{key}: {tasks[key]}") # Here f"..." is used to format the string
        elif choice == "3": # It will update the value of the key entered by the user to "Done" from "Not Done"
            taskdone=input("Enter task to mark done : ")
            tasks[taskdone]="Done"
        elif choice == '4':
            print("Logging out...")
            break # Exits out from the todo_list() function and control is now at the main_menu() function
        else:
            print("Invalid choice, please try again.")
main_menu() # Main menu function is called here
```
